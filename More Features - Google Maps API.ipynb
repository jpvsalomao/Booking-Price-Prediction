#### Function to extract relevant neighborhood information

```python
def neighborhood(lat, long, index):
    
    # API request for tourist attractions
    url_1 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+str(lat)+','+str(long)+'&radius=3200&types=tourist_attraction&key=AIzaSyB05iMkh9mZ2_m5Di5zgn400rAwnuRNFWc'
    response_1 = requests.get(url_1).json()
    
    # Create an empty list
    types = []
    
    # Save the types of places nearby
    for i in range(len(response_1['results'])):
        #print(i)
        types_1 = response_1['results'][i]['types']
        #print(types_1)
        types.append(types_1)
        
    # API request for airports
    url_2 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+str(lat)+','+str(long)+'&radius=3200&types=airport&key=AIzaSyB05iMkh9mZ2_m5Di5zgn400rAwnuRNFWc'
    response_2 = requests.get(url_2).json()
    
    # Save the types of places nearby
    for i in range(len(response_2['results'])):
        #print(i)
        types_2 = response_2['results'][i]['types']
        #print(types_2)
        types.append(types_2)
    
            
    # API request for night clubs
    url_3 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+str(lat)+','+str(long)+'&radius=3200&types=night_club&key=AIzaSyB05iMkh9mZ2_m5Di5zgn400rAwnuRNFWc'
    response_3 = requests.get(url_3).json()
    
    # Save the types of places nearby
    for i in range(len(response_3['results'])):
        #print(i)
        types_3 = response_3['results'][i]['types']
        #print(types_3)
        types.append(types_3)
        
    
    # API request for restaurants
    url_4 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+str(lat)+','+str(long)+'&radius=3200&types=restaurant&key=AIzaSyB05iMkh9mZ2_m5Di5zgn400rAwnuRNFWc'
    response_4 = requests.get(url_4).json()
    
    # Save the types of places nearby
    for i in range(len(response_4['results'])):
        #print(i)
        types_4 = response_4['results'][i]['types']
        #print(types_4)
        types.append(types_4)
        
        
    # API request for spas
    url_5 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+str(lat)+','+str(long)+'&radius=3200&types=spa&key=AIzaSyB05iMkh9mZ2_m5Di5zgn400rAwnuRNFWc'
    response_5 = requests.get(url_5).json()
    
    # Save the types of places nearby
    for i in range(len(response_5['results'])):
        #print(i)
        types_5 = response_5['results'][i]['types']
        #print(types_5)
        types.append(types_5)
        
    # Flatten the collated list of type tags
    flat_types = [item for sublist in types for item in sublist]
    counter=collections.Counter(flat_types)
    flat_types_dict = dict(counter)
    
    df_1 = pd.DataFrame(flat_types_dict,index=[index,])
    
    return df_1
```
